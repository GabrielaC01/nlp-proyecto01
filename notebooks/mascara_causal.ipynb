```python
import torch
import matplotlib.pyplot as plt
```

```python
def generar_mascara_causal(tamano):
    mascara = torch.triu(torch.ones(tamano, tamano), diagonal=1).bool()
    return mascara
```

```python
tamano = 10
mascara = generar_mascara_causal(tamano)

plt.figure(figsize=(6, 6))
plt.imshow(mascara, cmap="Greys")
plt.title("MÃ¡scara Causal (Triangular Inferior)")
plt.xlabel("Token atendido")
plt.ylabel("Token actual")
plt.colorbar()
plt.show()
```
